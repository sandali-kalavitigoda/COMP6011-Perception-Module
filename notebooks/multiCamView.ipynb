```python
# COMP6011_20908692
# python code for multiCamView
# upload images(front.jpg, left.jpg,right.jpg are provided to upload in the image folder in appendix)
from google.colab import files
files.upload()

import cv2
from google.colab.patches import cv2_imshow

# Load images uploaded
front = cv2.imread("front.jpg")
left = cv2.imread("left.jpg")
right = cv2.imread("right.jpg")

# Resize images
front = cv2.resize(front, (320,240))
left = cv2.resize(left, (320,240))
right = cv2.resize(right, (320,240))

# Combine multiple views
combined = cv2.hconcat([left, front, right])

cv2_imshow(combined)
cv2.imwrite("multiCamView.jpg", combined)
files.download("multiCamView.jpg")
cv2.waitKey(0)
cv2.destroyAllWindows()
```
