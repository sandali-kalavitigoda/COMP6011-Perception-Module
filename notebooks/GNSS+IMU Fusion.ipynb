```python
# COMP6011_20908692
# Python code for GNSS+IMU
import numpy as np
import matplotlib.pyplot as plt

n = 50
true_pos = np.linspace(0,10,n)

gnss = true_pos + np.random.normal(0,0.5,n)
imu = true_pos + np.cumsum(np.random.normal(0,0.2,n))

x, P = 0, 1
Q, R = 0.1, 0.5

est = []

for i in range(n):
    P += Q
    K = P/(P+R)
    x = x + K*(gnss[i]-x)
    P = (1-K)*P
    est.append(x)

plt.plot(true_pos,label="True")
plt.plot(gnss,label="GNSS")
plt.plot(imu,label="IMU")
plt.plot(est,label="Kalman",linewidth=2)
plt.legend()
plt.show()
```
